Verifica biblioteca de inicio

```python
def biblioteca(var):
  cabecalho = var.split('>')[0] #Esquerda
  cabecalho = cabecalho+'>'
  var = var.split('>')[-1] #Direita
  if "#include<stdio.h>" in cabecalho:
    print('cabeçalho Ok')
    return var
  else:
    return False
```

verifica função main.

```python
def main(var):
  main = var.split('{')[0]
  main += '{'
  var = var.split('{')[-1]
  if "intmain(){" in main:
    print('main Ok')
    return var
  else:
    return False
```

Busca Variáveis (levando em consideração o caso apontado, sem variações de tipo e escritas todas em mesma linha separadas por virgula)

```python
def buscaVariaveis(var):
  fun_main = var.split(';')[0] 
  variaveis2 = fun_main.split('int')[-1] #Remove o tipo das variavel(futuramente Type)
  
  variaveis2 = variaveis2.split(',')

  return variaveis2, var
```

Testa Scanf

```python
def testScanf(nTam,Token,variaveis):
  varPrintada = Token.split('&')[-1]
  print('Testeeeeeee: '+varPrintada)
  if ('scanf("%d",&' in Token[:12]) and (varPrintada.replace(")",'') in variaveis):
    print('Scanf correto: '+Token)
    return True;
```

Teste de Adição

```python
def testAdicao(Token,variaveis):
  somaVerifica = 0
  if ("=" in Token) and ("+" in Token): # c=a+b
    var = Token.split('=')[-1]
    var2 =  var.split('+')
    for x in range(len(var2)): 
      if var2[x] in variaveis:
        somaVerifica += 1 
      if somaVerifica == len(var2): #Ou seja, verificou todos
        print("A soma está correta: "+Token) 
        return True;  
```
